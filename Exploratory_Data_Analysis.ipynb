```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("train.csv")

df.head()
```

```python
df.shape                                
df.info()
```

```python

df.describe()
df.isnull().sum()
df.duplicated().sum()
df["Sex"].value_counts()
```

```python
df["Pclass"].value_counts()
df["Survived"].value_counts()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["Age"].dropna(), bins=20)

plt.title("Age Distribution")

plt.xlabel("Age")

plt.ylabel("Count")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df["Fare"])

plt.title("Fare Boxplot")

plt.savefig("images/boxplot.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(data=df, x="Age", y="Fare", hue="Survived")

plt.title("Age vs Fare")

plt.savefig("images/scatterplot.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.countplot(data=df, x="Sex", hue="Survived")

plt.title("Survival by Gender")

plt.savefig("images/survival_gender.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")

plt.savefig("images/heatmap.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
pair = sns.pairplot(
    df[["Age", "Fare", "Pclass", "Survived"]].dropna(),
    hue="Survived"
)

pair.savefig("images/pairplot.png", dpi=300)
```

```python
plt.figure(figsize=(6,5))

sns.countplot(data=df, x="Pclass", hue="Survived")

plt.title("Survival by Passenger Class")

plt.savefig("images/passenger_class.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
plt.figure(figsize=(6,5))

sns.countplot(data=df, x="Sex", hue="Survived")

plt.title("Survival by Gender")

plt.savefig("images/survival_gender.png", dpi=300, bbox_inches="tight")

plt.show()
```

Summary of Findings

• The dataset contains 891 passenger records and 12 features.
• Missing values were found in Age, Cabin, and Embarked.
• Most passengers were between 20–40 years old.
• The majority of passengers traveled in Third Class.
• Female passengers survived more frequently than males.
• First-class passengers had a higher survival rate.
• Ticket Fare contains several outliers.
• Fare has a positive relationship with Survival.
• Passenger Class strongly influences survival.
